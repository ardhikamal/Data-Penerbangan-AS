Menghilangkan Outliers

Dibawah ini merupakan distribusi dan rangkuman statistik untuk kolom DepDelay dan ArrDelay

```python
# fungsi untuk menampilkan rangkuman statistik dan distribusi untuk sebuah kolom
def show_distribution(var_data):
    from matplotlib import pyplot as plt

# Mendapatkan data statistik
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]

    print(var_data.name,'\nMinimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

# Membuat gambar untuk 2 subplot(2 Baris, 1 kolom)
    fig, ax = plt.subplots(2, 1, figsize = (10,4))
# menggambar histogram
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')

# menambahkan garis untuk mean, median, dan modus
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

# menggambar plot kotak   
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Value')

# menambah judul ke gambar
    fig.suptitle(var_data.name)

# memanggil fungsi untuk masing-masing bidang keterlambatan
delayFields = ['DepDelay','ArrDelay']
for col in delayFields:
    show_distribution(df_flights[col])
```

Ada beberapa outlier di ujung bawah dan atas dari kedua variable, namun kebanyakan pada ujung atas.

Data akan dipangkas sehingga kita hanya menyertakan baris di mana nilai untuk bidang ini berada dalam persentil ke-1 dan ke-

90.

```python
# Memangkas outliers untuk ArrDelay berdasarkan persentil 1% and 90%
ArrDelay_01pcntile = df_flights.ArrDelay.quantile(0.01)
ArrDelay_90pcntile = df_flights.ArrDelay.quantile(0.90)
df_flights = df_flights[df_flights.ArrDelay < ArrDelay_90pcntile]
df_flights = df_flights[df_flights.ArrDelay > ArrDelay_01pcntile]

# Memangkas outliers untuk DepDelay berdasarkan persentil 1% and 90%
DepDelay_01pcntile = df_flights.DepDelay.quantile(0.01)
DepDelay_90pcntile = df_flights.DepDelay.quantile(0.90)
df_flights = df_flights[df_flights.DepDelay < DepDelay_90pcntile]
df_flights = df_flights[df_flights.DepDelay > DepDelay_01pcntile]

# Memperlihatkan distribusi yang telah direvisi
for col in delayFields:
    show_distribution(df_flights[col])
```

Bandara keberangkatan mana yang memiliki nilai keterlambatan keberangkatan tertinggi?

```python
departure_airport_group = df_flights.groupby(df_flights.OriginAirportName)

mean_departure_delays = pd.DataFrame(departure_airport_group['DepDelay'].mean()).sort_values('DepDelay', ascending=False)
mean_departure_delays.plot(kind = "bar", figsize=(12,12))
mean_departure_delays
```

Apakah keterlambatan keberangkatan menyebabkan keterlambatan kedatangan yang lebih lama daripada kedatangan tepat waktu?

```python
df_flights.boxplot(column='ArrDelay', by='DepDel15', figsize=(8,8))
```

Rute yang mana (dari bandara awal ke bandara tujuan) yang memiliki kedatangan yang paling terlambat?

```python
# Menambahkan kolom rute
routes  = pd.Series(df_flights['OriginAirportName'] + ' > ' + df_flights['DestAirportName'])
df_flights = pd.concat([df_flights, routes.rename("Route")], axis=1)

# Pengelompokan berdasarkan rute
route_group = df_flights.groupby(df_flights.Route)
pd.DataFrame(route_group['ArrDel15'].sum()).sort_values('ArrDel15', ascending=False)
```

Rute mana yang memiliki rata-rata keterlambatan kedatangan yang tertinggi?

```python
pd.DataFrame(route_group['ArrDelay'].mean()).sort_values('ArrDelay', ascending=False)
```
